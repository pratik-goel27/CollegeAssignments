```python
# QUESTION-1
def replace_repetive_char(str):
    result = str[0]
    for i in range(1,len(str)):
        if str[i] == str[i-1]:
            result += '*'
        else:
            result += str[i]
    return result
print(replace_repetive_char('balloon'))
```

```python
# QUESTION-6
def replace_str(str):
    vowels = "aeiou"
    result = ""
    for c in str:
        if c in vowels:
            result += vowels[(vowels.index(c)+1)%5]
        else:
            result += c
    return result
print(replace_str("aeiou"))
```
