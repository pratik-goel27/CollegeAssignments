```python
# def mode(arr):
#     values, counts = np.unique(arr, return_counts=True)
#     values[np.argmax(counts)]


def median(arr):
    return np.median(arr)

def mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]


C = np.array([[[7, 8, 9],
               [7, 7, 6]],
              [[5, 7, 6],
               [7, 9, 5]]])

print("Array C:\n", C)
print("Median:", median(C))
print("Mode:", mode(C))
```
