```python
lst = [2**i for i in range(6)]
arr = np.reshape(lst, (2,3))
print(arr)
f = arr.flatten()
print("FLATTEN: \nflattened arr :\n",f)
f[0] = 100
print("after modification :\n",f)
print("orginal arr after mod: \n",arr)

r = arr.ravel()
print("\nRAVEL: \nraveled arr :\n",r)
r[0] = 100
print("after modification :\n",r)
print("orginal arr after mod: \n",arr)
```
