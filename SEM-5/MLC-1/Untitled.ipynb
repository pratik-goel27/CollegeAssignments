```python
#Q1 :-
import numpy as np
arr = np.arange(1,21)
primes = []
for i in arr:
    if i > 1:
        flag = True
    for j in range(2,int(i**0.5) + 1):
        if i % j == 0:
            flag = False
            break
    if flag:
        primes.append(int(i))

mean = np.mean(primes)
var = np.var(primes)
print(arr)
print(primes)
print(mean)
print(var)
```

```python
# Q10
data = pd.DataFrame({
    'Name': ['AC', 'Fan', 'Ceiling Lights', 'Inchi Tape','Screws and nuts'],
    'Quantity': [2,4,10,2,200],
    'Price': [100,200,300,400,500]
    })

data['Total'] = data['Quantity'] * data['Price']
print(data,"\n")
max_prod = data.loc[data['Total'].idxmax()]
print("Maximum sales revenue: ",max_prod['Name'])
```

```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['pratik','ankush','aditya','debasish','prasanjit','saurav'],
    'Age': [20,19,18,20,23,21],
    'Marks': [90,80,90,80,70,60]
})
avg = df['Marks'].mean()
print(df,"\n")
print(avg)

print("Above avg:\n", df[df['Marks']>avg])
shortlisting = df[(df['Age'] < 20) & (df['Marks'] > 60)]
print("\n",shortlisting['Name'])
```
