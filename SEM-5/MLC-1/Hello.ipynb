```python
#Array with values from 1-20 :- (1st way)
arr = np.array([], dtype = int)
for i in range(1, 21):
    arr = np.append(arr, i)

print(arr)

#(Using arange() method) :-
arr = np.arange(1,21)
print(arr)
```

```python
temp = np.array([], dtype = int)
#Finding the Prime No's :-
for i in range(len(arr)):
    if i < 2: continue
    flag = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            flag = False
            break
    if flag:
        temp = np.append(temp, i)

#Calc Mean:-
sum = 0
for i in range(len(temp)):
    sum += temp[i]

mean = sum / len(temp)
print(mean)
#Calc Variance :-
variance = np.var(temp)
print(variance)
```
