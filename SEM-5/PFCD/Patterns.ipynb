```python
for i in range(1,5):
    for j in range(1,i+1):
        print("*", end = " ")
    print("\n")
```

```python
n = 4
for i in range(n):
    for j in range(n-1-i):
        print(end=" ")
    for j in range(i+1):
        print("*",end="")
    print("\n")
```

```python
ch = 'A'
for i in range(8):
    for j in range(i):
        print(ch, end=" ")
        ch = chr(ord(ch) + 1)
    print("\n")
```
